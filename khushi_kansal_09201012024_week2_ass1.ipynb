```python
#Q11
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

label_cols = [
    "Gender",
    "SubscriptionType",
    "FavoriteGenre",
    "SubscriptionRenewed"
]

encoders = {}

for col in label_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    encoders[col] = le

print("\nEncoded Dataset:")
print(df_encoded.head())

```

```python
#Q12
X = df_encoded.drop("SubscriptionRenewed", axis=1)
y = df_encoded["SubscriptionRenewed"]

print("\nFeature Matrix Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
#Q14
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("\nTraining Samples:", len(X_train))
print("Testing Samples:", len(X_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)

print("\nDecision Tree Model Trained")
```

```python
from sklearn.metrics import accuracy_score

dt_pred = dt_model.predict(X_test)

dt_accuracy = accuracy_score(y_test, dt_pred)

print("\nDecision Tree Accuracy:")
print(round(dt_accuracy*100,2), "%")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, dt_pred)

print("\nConfusion Matrix:")
print(cm)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

knn_pred = knn.predict(X_test)

print("\nKNN Model Trained")
```

```python
knn_accuracy = accuracy_score(y_test, knn_pred)

print("\nDecision Tree Accuracy:",
      round(dt_accuracy*100,2), "%")

print("KNN Accuracy:",
      round(knn_accuracy*100,2), "%")

if knn_accuracy > dt_accuracy:
    print("KNN Performs Better")
else:
    print("Decision Tree Performs Better")
```

```python
from sklearn.linear_model import LinearRegression

X_reg = df_encoded.drop("MonthlySpend", axis=1)
y_reg = df_encoded["MonthlySpend"]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.20,
    random_state=42
)

lr = LinearRegression()

lr.fit(X_train_reg, y_train_reg)

print("\nLinear Regression Model Trained")
```

```python
sample_user = X_reg.iloc[[0]]

prediction = lr.predict(sample_user)

print("\nPredicted Monthly Spend:")
print(prediction[0])
```

1. Which factors appear to influence subscription renewal the most?

2. Why is subscription renewal a classification problem?

Subscription renewal is a classification problem because the target variable has only two possible outcomes: Yes or No. The goal of the model is to classify each user into one of these categories based on their demographic information, viewing habits, and spending behavior. Since the output is categorical rather than numerical, classification algorithms such as Decision Tree and KNN are appropriate.

3. Why is monthly spending a regression problem?

Monthly spending is a regression problem because the value to be predicted is numerical and continuous. A user's spending can vary across a wide range of amounts rather than belonging to fixed categories. The objective is to estimate the exact spending amount based on user characteristics, which makes Linear Regression a suitable technique.

Based on the accuracy scores obtained during model evaluation, the algorithm with the higher accuracy performed better for subscription renewal prediction. In this dataset, KNN achieved a higher accuracy than the Decision Tree model, indicating that it was able to classify users more effectively. Therefore, KNN can be considered the better-performing model for this particular dataset.
